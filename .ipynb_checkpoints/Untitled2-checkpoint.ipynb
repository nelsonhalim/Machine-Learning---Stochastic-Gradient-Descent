```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math as mt


#TABEL 100 - TUGAS 1

data = pd.read_csv('irisdata.csv', header=None, nrows=100)

data[4] = data[4].apply(lambda x: str(x).replace('Iris-setosa', '1'))
data[4] = data[4].apply(lambda x: str(x).replace('Iris-versicolor', '0'))
data[4] = data[4].astype('int64')

#TABEL TRAINING - TUGAS 2

tabel_1_train = pd.read_csv('irisdata.csv', header=None, nrows=40)
tabel_0_train = pd.read_csv('irisdata.csv', header=None, skiprows=50, nrows=40)

tabel_1_train[4] = tabel_1_train[4].apply(lambda x: str(x).replace('Iris-setosa', '1'))
tabel_1_train[4] = tabel_1_train[4].astype('int64')

tabel_0_train[4] = tabel_0_train[4].apply(lambda x: str(x).replace('Iris-versicolor', '0'))
tabel_0_train[4] = tabel_0_train[4].astype('int64')

tabel_train= pd.concat([tabel_1_train,tabel_0_train])
tabel_train = tabel_train.reset_index(drop=True)

# #TABEL VALIDASI

tabel_1_validasi = pd.read_csv('irisdata.csv', header=None, skiprows=39, nrows=10)
tabel_0_validasi = pd.read_csv('irisdata.csv', header=None, skiprows=90, nrows=10)


tabel_1_validasi[4] = tabel_1_validasi[4].apply(lambda x: str(x).replace('Iris-setosa', '1'))
tabel_1_validasi[4] = tabel_1_validasi[4].astype('int64')

tabel_0_validasi[4] = tabel_0_validasi[4].apply(lambda x: str(x).replace('Iris-versicolor', '0'))
tabel_0_validasi[4] = tabel_0_validasi[4].astype('int64')

tabel_validasi = pd.concat([tabel_1_validasi,tabel_0_validasi])
tabel_validasi = tabel_validasi.reset_index(drop=True)


type(tabel_validasi)
```

```python
#INISIASI
alpha = 0.1

teta = np.array([0.1,0.15,0.2,0.3])

pteta= pd.DataFrame(teta)

bias = 0.4

d_bias = 0

h = 0

array_dteta = np.empty(4)

total_error_semua_train = np.zeros(60)

total_error_semua_validasi = np.zeros(60)

local_error_train = 0

local_error_val= 0


```

```python
# by def
def H(x, teta, bias, a):
    return np.dot(x.iloc[a, :4], np.transpose(teta)) + bias


def sigmoid(h):
    return 1 / (1 + mt.exp(-1 * h))


def error(tabel,sigmoid, a):
    return (tabel.iloc[a, 4] - sigmoid ) ** 2


def prediksi(sigmoid):
    if sigmoid > 0.5:
        return 1
    else:
        return 0


def d_teta(sigmoid, fakta, array_x, i):
    return 2 * (sigmoid - fakta) * (1 - sigmoid) * sigmoid * array_x[i]


def d_bias(sigmoid, fakta):
    return 2 * (sigmoid - fakta) * (1 - sigmoid) * sigmoid


def teta_baru(teta, alpha, d_teta, i):
    return teta[i] - (alpha * d_teta[i])


def bias_baru(bias, alpha, d_bias):
    return bias - (alpha * d_bias)


```

```python
# EPOCH:60
for m in range(60):

    # TRAIN
    # ROW:80
    for i in range(80):

        # print('')
        # print('bias: ',bias)
        # print('teta: ',teta)


        array_x = np.array(tabel_train.iloc[i, :4])
        fakta = tabel_train.iloc[i, 4]

        h_rumus = H(tabel_train, teta, bias, i)

        sigmoid_rumus = sigmoid(h_rumus)

        error_rumus = error(tabel_train,sigmoid_rumus, i)
        local_error_train = local_error_train + error_rumus

        prediksi_rumus = prediksi(sigmoid_rumus)

        for j in range(len(array_x)):
            array_dteta[j] = d_teta(sigmoid_rumus, fakta, array_x, j)

        d_bias_rumus = d_bias(sigmoid_rumus, fakta)

        for j in range(len(teta)):
            teta[j] = teta_baru(teta, alpha, array_dteta, j)

        bias_baru_rumus = bias_baru(bias, alpha, d_bias_rumus)

        bias = bias_baru_rumus

        # #LIHAT TRAIN
        # print('____________________________')
        # print('ROW: ',i)
        # print('')
        #
        # print('alpha: ',alpha)
        # print('array x: ', array_x)
        # print('fakta x: ', fakta)
        # print('h_rumus: ', h_rumus)
        # print('sigmoid_rumus: ', sigmoid_rumus)
        # print('error_rumus: ', error_rumus)
        # print('prediksi: ',prediksi_rumus)
        # print('array_dteta: ', array_dteta)
        # print('d_bias: ', d_bias_rumus)
        # print('teta baru: ',teta)
        # print('bias baru: ', bias)
        # print('')
        # print('local_error_train: ', local_error_train)
        # print('')
        # print('____________________________________')


    total_error_semua_train[m] = local_error_train
    local_error_train = 0


    # VALIDASI
    # ROW:20

    for i in range(20): #len(tabel_validasi)

        array_x = np.array(tabel_validasi.iloc[i, :4])
        fakta = tabel_validasi.iloc[i, 4]

        h_rumus = H(tabel_validasi, teta, bias, i)

        sigmoid_rumus = sigmoid(h_rumus)

        error_rumus = error(tabel_validasi,sigmoid_rumus, i)
        local_error_val = local_error_val + error_rumus

        prediksi_rumus = prediksi(sigmoid_rumus)

        # #LIHAT VALIDASI
        # print('____________________________')
        # print('ROW: ',i)
        # print('')
        #
        # print('alpha: ',alpha)
        # print('array x: ', array_x)
        # print('fakta x: ', fakta)
        # print('h_rumus: ', h_rumus)
        # print('sigmoid_rumus: ', sigmoid_rumus)
        # print('error_rumus: ', error_rumus)
        # print('prediksi: ',prediksi_rumus)
        # print('array_dteta: ', array_dteta)
        # print('d_bias: ', d_bias_rumus)
        # print('teta baru: ',teta)
        # print('bias baru: ', bias)
        # print('')
        # print('local_error_val: ', local_error_val)
        # print('')
        # print('____________________________________')


    total_error_semua_validasi[m] = local_error_val
    local_error_val = 0

```

```python
#PLOT
plt.plot(total_error_semua_train, color="blue")
plt.plot(total_error_semua_validasi, color= "red")
plt.title("Grafik Error Setiap Epoch")
plt.legend(["Data Training", "Data Validation"])
plt.ylabel('error')
plt.xlabel('epoch')

plt.show()
```
